```python
import os

import mlflow
import torch
import pandas as pd
import torch.nn as nn
import numpy as np

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
mlflow_uri = os.environ['MLFLOW_URI']
```

```python
datum=[]
for year in range(2017,2022):
    data=pd.read_csv('data/'+str(year)+'_seoul.csv', encoding='cp949').iloc[:,[2,3]]
    datum.append(data)
```

```python
data=pd.concat(datum, ignore_index=True)
```

```python
data.columns=['date','temp']
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data=data.fillna(method='ffill')
```

```python
data.isnull().sum()
```

```python
div=23
split_data=[]
scaled_data=[]
scaler=MinMaxScaler()
for idx in range(0,len(data),div):
    temp_data=data.iloc[idx:idx+div,:]
    split_data.append(temp_data)
    temp_data.iloc[:,[1]]=scaler.fit_transform(temp_data.iloc[:,[1]])
    scaled_data.append(temp_data)
```

```python
scaled_data
```

```python
train_ratio=0.6
test_ratio=0.2
validataion_ratio=0.2

train_set, test_set=train_test_split(scaled_data, test_size=test_ratio)
train_set, validation_set=train_test_split(train_set, test_size=validataion_ratio/(train_ratio+test_ratio))
len(train_set), len(validation_set), len(test_set)
```

```python
len(train_set), len(validation_set), len(test_set)
device = torch.device('cpu')
```

```python
class VanillaRNN(nn.Module):
    def __init__(self, input_size, output_size, hidden_dim, n_layers):
        super(VanillaRNN, self).__init__()
        
        self.hidden_dim = hidden_dim
        self.n_layers = n_layers
        
        self.rnn = nn.RNN(input_size, hidden_dim, n_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_size)
        
    def forward(self, x, hidden=None):
        
        if hidden is None:
            batch_size = x.size(0)
            hidden =self.init_hidden(batch_size)
        out, hidden = self.rnn(x, hidden)
        
        out = out.contiguous().view(-1, self.hidden_dim)
        out = self.fc(out)
        
        return out, hidden
    
    def init_hidden(self, batch_size):
        hidden = torch.zeros(self.n_layers, batch_size, self.hidden_dim).to(device)
        return hidden
    
    def predict(self, input_data):
        result=[]
        model.eval()
        with torch.no_grad():
            __, hidden = model(input_data[:,0,:,:])
        
            for i in range(1, 12):
                output, hidden = model(input_data[:,i,:,:], hidden)

            for i in range(11):
                output = output.reshape(-1,1,1)
                output, hidden = model(output, hidden)
                result.append(output)
                
        return result
```

```python
mlflow.set_tracking_uri(mlflow_uri)
mlflow.set_experiment('vanilla-RNN')
```

```python
hidden_dim=48
n_layers=4
model = VanillaRNN(input_size=1, output_size=1, hidden_dim=hidden_dim, n_layers=n_layers)
model = model.to(device)
n_epochs=100
lr=0.01
batch_size=128
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
```

```python
start, end = 0, 32
batch_data = train_set[start:end]
input_data = np.array([list(batch['temp'][:12]) for batch in batch_data])
output_data = np.array([list(batch['temp'][12:]) for batch in batch_data])
```

```python
np.array(output_data).shape
```

```python
for epoch in range(1, n_epochs+1):
    for idx in range(0, len(train_set), batch_size):
        try:
            batch_data = train_set[idx:idx+batch_size]
            input_data = torch.Tensor(np.array([list(batch['temp'][:12]) for batch in batch_data]))
            output_data = torch.Tensor(np.array([list(batch['temp'][12:]) for batch in batch_data]))
            input_data = input_data.reshape(-1, 12, 1, 1)
            output_data = output_data.reshape(-1,11,1, 1)
            optimizer.zero_grad()
            loss = 0
            __, hidden = model(input_data[:,0,:,:])
        
            for i in range(1, 12):
                output, hidden = model(input_data[:,i,:,:], hidden)

            for i in range(11):
                output = output.reshape(-1,1,1)
                output, hidden = model(output, hidden)
            
                loss += criterion(output.reshape(-1),output_data[:,i,:,:].reshape(-1))
        
            training_loss = loss
            loss.backward()
            optimizer.step()
        except:
            pass
    if epoch%5==0:
        print("epoch:", epoch, "TRAINING LOSS:", loss.item())
        
    if epoch%10 == 0:
        try:
            
            for idx in range(0, len(validation_set), batch_size):
                loss = 0
                batch_data = validation_set[idx:idx+batch_size]
            
                input_data = torch.Tensor(np.array([list(batch['temp'][:12]) for batch in batch_data]))
                output_data = torch.Tensor(np.array([list(batch['temp'][12:]) for batch in batch_data]))
                input_data = input_data.reshape(-1, 12, 1, 1)
                output_data = output_data.reshape(-1,11,1, 1)
                
                __, hidden = model(input_data[:,0,:,:])
        
                for i in range(1, 12):
                    output, hidden = model(input_data[:,i,:,:], hidden)

                for i in range(11):
                    output = output.reshape(-1,1,1)
                    output, hidden = model(output, hidden)
                    loss += criterion(output.reshape(-1),output_data[:,i,:,:].reshape(-1))
            validation_loss = loss
            print("epoch:", epoch, "VALIDATION LOSS:", loss.item())
        except:
            print(batch_data)
```

```python
loss = 0
for idx in range(0, len(test_set), batch_size):
    try:
        batch_data = test_set[idx:idx+batch_size]
            
        input_data = torch.Tensor(np.array([list(batch['temp'][:12]) for batch in batch_data]))
        output_data = torch.Tensor(np.array([list(batch['temp'][12:]) for batch in batch_data]))
        input_data = input_data.reshape(-1, 12, 1, 1)
        output_data = output_data.reshape(-1,11,1, 1)
        
        __, hidden = model(input_data[:,0,:,:])
        
        for i in range(1, 12):
            output, hidden = model(input_data[:,i,:,:], hidden)

        for i in range(11):
            output = output.reshape(-1,1,1)
            output, hidden = model(output, hidden)
            loss += criterion(output.reshape(-1),output_data[:,i,:,:].reshape(-1))
    
    except:
        print('')
test_loss = loss
print("TEST LOSS:", loss.item())        
```
